```python
// Program to find the kth root of a number in logn time complexity

// importing the libraries
#include<iostream>
#include<cstdio>
#include <boost/multiprecision/cpp_int.hpp>
using namespace boost::multiprecision;
using namespace std;

// function to compute the power of the element i.e x^y
cpp_int powerofx(cpp_int x, cpp_int y) {
    cpp_int xpy;
    // x is the number and y is power to x 
    if (y == 0) {
        return 1;
    }
    cpp_int temp;
    temp = powerofx(x, y / 2);
    if (y % 2 == 0) {
        xpy = temp * temp; // when the power is even 
    }
    else {
        xpy = x * temp * temp; // when the power is odd
    }
    return xpy;
}

// function to compute kth root of n i.e n^(1/k)
cpp_int kthRootOfn(cpp_int l, cpp_int h, cpp_int n, cpp_int k) {
    cpp_int m;
    cpp_int mp;
    cpp_int mnp;
    m = (l + h) / 2;
    if (l <= h) { // function doesn't work for l>h as it is invalid case 
        mp = powerofx(m, k); // kth power of the mid element 
        mnp = powerofx(m + 1, k);//kth power of the mid+1 element
        if (mp <= n) {
            if (mnp > n) {
                return m; // return the mid element the kth root if its kth power is less than or equal to n and mid+1 to the power k is greater than n
            }
            else {
                return kthRootOfn(m + 1, h, n, k); // search in half space after mid as k th power is not between 0 to mid
            }
        }
        else {
            return kthRootOfn(0, m - 1, n, k);  // search between 0 to mid-1 as mid power k is greater than n and we need not search in other half from mid+1 to end
        }
    }
}
//main function
int main() {
    int t;
    cin >> t; // getting the input test case 
    while (t > 0) {
        cpp_int n, k;
        cpp_int result;
        //enter the numbers n and k
        cin >> n;
        cin >> k;
        //cout<<n<<" "<<k<<"\n";
        result = kthRootOfn(0, n, n, k); // called the function and the value returned by the function will be stored in variable result
        cout << result << "\n";
        t = t - 1; //decrement the test case number as the result for above test case is computed by now
    }

}
```
